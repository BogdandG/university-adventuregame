```python
print("""
DocString:
    
    A) Introduction:
    
    This text adventure game was inspired by popular Chinese novel and MMORPG game
    Jade Dynasty or Zhu Xian. Couple of series and films exists based on that novel.
    However, this episode was fully written by myself and not exactly correspond to the novel scenes.
    
    Places of actions:
    1:Game start
    2:Abode 
    3:Pit
    4:Garden 
    5: Game end 
    
    Additional functions:
    Clan selection 

    B) Known Bugs and/or Errors:
    None.
    
    **** ReadMe **** 
    C) Additional information:
    
     * Due to the time limitations, excess additional text of the story was highlihted, 
    in order to create a shortened version to suit the time. 
    Please don't read the highlighted text during the official time measuring.
    
    Highlighted text is present in this format: \033[0;30;47m'text'\033[m
    
    
    * Sleep function could be used more mindfully in order to obtain more smoother user expirience, 
    however due to time limit, it was decided to keep low value of sleep functions and couple of big strings to save the time.
    
   
    * Tested playing time for wining trial:
        Short version: 3.48m
        Full version: 7m

"""
     )
```

```python
from time import sleep
from sys import exit
import random
from random import randint
red_text = '\033[1;31;1m'
highlighted = '\033[0;30;47m'
reset = '\033[m' 

def game_start():
    global character_name
    #Creating character name
    character_name = input(prompt = "Please enter the character name to proceed:\n>").lower()
    if not character_name:
        print("<Please enter at least your nickname>")
        sleep(1)
        character_name = input().lower()
    
    #Printing introduction to the story
    print(f"\nSince the very dawn of the world, humanity has watched a strange and hostile environment around them.{highlighted} What they saw  led them to believe that divine beings inhabited the nine heavens above, while demons occupied the vastitude of hell. {reset}")
    sleep(1.5)
    print("Due to weaknesses intrinsic to mankind, power was desired by everyone in pursuit of immortality – a rather elusive   goal.")
    sleep(1.5)
    print(f"""Thus, two conflicting clans were born: \n\nThe Sky clan  \nThe Evil clan.\n""")
    sleep(1.5)
    print(f"""{highlighted}Legends say that the founders of the clans possessed the knowledge that was key to immortality. 
Only the most excelling clan members could have a chance at the knowledge being passed to them...""")
    sleep(2)

    print(f"""{highlighted}\nOr maybe it is a mere legend – you, {red_text}{character_name.capitalize()}{reset}{highlighted}, will have to see it for yourself.{reset}
{'-'*117}""")
    sleep(3)
    
    #Progressing to clan selection
    input(prompt='Press enter to continue\n')
    clan_selection()
    sleep(3)
    
    #Beginning of the journey 
    print(f"""\nSo let your journey begin, {red_text}{character_name.capitalize()}{reset}!
To earn the clan's trust, you need to complete a special quest from the Master Mentor and pass the initiation.
""")
    sleep(3)
    print("One night, unexpectedly, you receive a mission scroll from your master.")
    scroll = input(prompt='Press enter to read the scroll\n')
    print("Scroll unfolded:\n")
    print(f"""<<<Head to the Lunar abode, \n\tfind the ancient tome hidden years go,\n\t\tpass the quest to join the clan, \n\t\t\t\tYour Master of {clan}!>>>""")
   
    #Proceeding to first room 
    print(f"{'-'*117}")
    input(prompt='Press enter to continue\n')
    abode()
    
    
def clan_selection():
    #Setting clan choice variables
    good_clan = "1) Sky clan"
    evil_clan = "2) Evil clan"
    #Selecting a clan 
    print(f"Please select the clan to proceed:\n{good_clan}\n{evil_clan}")
    global clan
    global opposite_clan 
    clan = input(prompt='>').lower()
    
    
    #Creating clan choise based on user input + clan information + setting chosen clan as global variable. 
    if clan in good_clan:
        print("""\nOn the seven great peaks, surrounded by clouds, settled the Sky clan.""")
        sleep(0.5)
        print(f"{highlighted}Legends say that the founder of the clan was a spiritual master with special powers. ")
        sleep(0.5)
        print("He has chosen and trained 10 of the best disciples, who would be ruling on his deathbed.")
        sleep(0.5)
        print('For the next hundreds of years after the death of the holy master, instead of prospering, the Sky clan went through \ngreat decline. ')
        sleep(0.5)
        print(f"""Still, it's not clear, if this was due to the will of heaven or the lack of faith in the disciples.{reset}
{'-'*117}""")
    #setting new global variable for clan + elimination of number for future use
        clan  = good_clan.strip("1) ")
        opposite_clan = evil_clan.strip("2) ")
        
     #Creating clan choise based on user input + clan information + setting chosen clan as global variable.   
    elif clan in evil_clan:
        print(f"""\nUnder the red clouds on the peak the Evil clan settled. The founder of the clan was an old monk who was expelled from the Sky clan for his thirst for power. """)
        sleep(0.5)
        print(f'{highlighted}The exile took up against people and stepped on the evil path, killing the innocent.' )
        sleep(0.5)
        print('Over time, his heart was shrouded in shadow, and his eyes had a strange red sheen.' )
        sleep(0.5)
        print(f'''But he received the cherished power that can destroy the world.{reset}
{'-'*117}''')
    #setting new global variable for clan + elimination of number for future use 
        clan  = evil_clan.strip("2) ")
        opposite_clan = good_clan.strip("1) ")
    #Debugging     
    else:
        sleep(2)
        print("Hurry up, apprentice, choose wisely - great future awaits you")
        clan_selection()
        
        
#First interactive room in the story         
def abode():
    print("\nYour journey has been, long and thorny...")
    sleep(1)
    print("\nUpon reaching the cave, you noticed a dim fire spark hovering above the ground not far from the entrance.\n")
    #Two possible choices for the player
    abode_choise_1 = '1) Continue with the quest and enter the Lunar abode'
    abode_choise_2 = '2) Follow and catch the spark'
    
    #creating a while loop with nested condition for progression 
    proceed = 0
    while proceed <1:
        print(f"Please select: \n{abode_choise_1} \n{abode_choise_2}")
        #Asking for user input
        choise_abode = input("> ")
        
        # Failure choise
        if choise_abode.lower() in abode_choise_1.lower():
            print("You entered the Lunar abode")
            sleep(1)
            print("While going deeper inside the abode, you accidently stepped on the floor slab and activated some mechanism.")
            sleep(1)
            print("A deafening roar came from nowhere, and the cave began to collapse.")
            sleep(3)
            print("Apparently, there was another way to enter the abode, however you will no be able to find it out...")
            sleep(3)
            fail()
        # ****** Abode progression choice******
        elif choise_abode in abode_choise_2.lower():
            print("As soon as you approached the spark and tried to touch it.")
            sleep(1)
            print(f"It lit up with a bright scarlet color and began to soar towards the forest, enticing you to follow.")
            sleep(1)
            print(f"{highlighted}\nSuccumb to temptation, you followed strange spark up to the glade in the forest.{reset}")
            sleep(1)
            print(f"Reaching the glade, a spark instantly disappeared, leaving you alone.")
            sleep(1)
            print("\nJust as you were about to head back to the cave, the ground began to crack under your feet, forming large pit.")
            sleep(1)
            print("Clinging to old roots and grass, you sank into the ground.")
            proceed+=1
            sleep(1)
            print(f"{'-'*117}") 
            pit()
        #Debugging
        else:
            print(f"Time is running, {red_text}{character_name.capitalize()}{reset}!! You sure you want to become a disciple?\n\nHint: please use numbers to proceed")
        
#Second room, proceeding with the story
def pit():
    #Entering the second location
    input(prompt = "Press enter to wake up\n")
    print("When you woke up, you found yourself on the cold ground in a dark place. \nKnowing there will not be any help, you try to tread carefully and slowly move forward.")
    sleep(1)
    print("\nSuddenly, the same spark appeared in front of you.")
    sleep(1)
    print(f"{highlighted}Bright light blinded your eyes,\n'Maybe it can help me?' - you thought.{reset}")
    #Choosing options to proceed with the quest
    pit_choice_1 = "1) Find exit by yourself"
    pit_choice_2 = "2) Follow the spark"
    proceed = 0
    while proceed < 1:
        print(f"\nPlease select what you want to do: \n{pit_choice_1} \n{pit_choice_2}")
        choice_pit = input("\n>")
        #Pit fail choise
        if choice_pit.lower() in pit_choice_1.lower():
            print("""You entered the first tunnel and went ahead. 
Shortly you realized, that the cave consists of many tunnels and \n\n\tmost likely you will wander in them forever...""")
            sleep(1)
            fail()
        # ******Pit progression choice 1 ********
        elif choice_pit.lower() in pit_choice_2.lower():
            print(f"{highlighted}It was your only hope to get out of the cave...{reset}")
            sleep(1)
            print("You obediently followed the spark until you saw how the moonlight began to illuminate the cave.")
            sleep(1)
            print(f"{highlighted}Your gaze fell on the old altar, which was covered with dust and spiderwebs.{reset}")
            sleep(1)
            print("A blood-red medallion layed on the altar, with a huge stone shimmering in the moonlight...")
            proceed+=1
            print(f"{'-'*117}") 
            
        #Debugging
        else:
            print(f"Time is running, {red_text}{character_name.capitalize()}{reset}!! You sure you want to become a disciple?\n\nHint: please use numbers to proceed")
        
    
    pit_choice_3 = "1) Take the medallion"
    pit_choice_4 = "2) Leave the medallion and search for exit"
        
    while proceed < 2:
        print(f"\nPlease select what you want to do: \n{pit_choice_3} \n{pit_choice_4}")
        choice_pit_medallion = input(">")
        # ******Pit progression choice 2 ********
        if choice_pit_medallion.lower() in pit_choice_3.lower():
            print("You touched the medallion")
            sleep(1)
            print(f"{highlighted}Strong wind arose in the cave! \n\nThe deafening roar of the dragon was heard in the mountains and a bright scarlet light hit the sky illuminating the entire forest.{reset} ")
            sleep(1)
            print(f"""\n\nBright light surrounded you and you heard a voice whispering: \n{'*'*117}\n
            \t\t'You are the chosen one, now, you are the gods killer.'\n{'*'*117}""")
            proceed +=1
        # Alternate dead-end choice
        elif choice_pit_medallion.lower() in pit_choice_4.lower():
            print("You didn't find an exit! You do not know what to do, \n\n\tmaybe it is worth taking a medallion?")
            input(prompt = "Press enter to continue\n")
        #Debugging 
        else:
            print(f"Time is running, {red_text}{character_name.capitalize()}{reset}!! You sure you want to become a disciple?\n\nHint: please use numbers to proceed")
     
    # Storyline progression 
    print("""\n'Am I the gods killer?'\n
    'But what does it mean?' \n
    \t'What should I do now?'\n
    \t\t\tthese thoughts haunted your mind.""")
    
    sleep(6)
    print(f"""{highlighted}\nStrange voice fell silent, and that bright light disappeared.\n
The sprak actively stirred and continued to fly, \n\n"I should follow it ..." you thought!{reset}
""")
    sleep(3)
    ####################################################
    #Creation of first interactive game 
    #Initially an idea of tic tac toe game  for the particular episode was developed
    #However due to time limit and class object restriction a simple guessing game was selected, modified for the environment.
    ####################################################
    print(f"""\nSuddenly the light stopped.\n
{'-'*117}""")
    input(prompt='Press enter to continue\n')
    print(f"""\nThere was an old door ahead, blocking the way. 
Approaching the door, you found three strange stone cups, they flew into the air and the spark hid inside one of them.""")
    sleep(2)
    print("Cups slowly went down, and suddenly began to mix.")
    sleep(2)
    print("'I should guess the right one!'")
    
   #Chances to proceed
    chances = 6

    left = "1) Left"
    middle = "2) Middle"
    right = "3) Right"

    while chances > 0:

        #Choosing the cup
        print(f"""Please choose the cup with the spark: \n{left} \n{middle} \n{right}
    """)
        player_stone_cup_input = input(prompt="> ")

        #Processing the users choice 
        if player_stone_cup_input.lower() in left.lower():
            cup_choice = 1 
        elif player_stone_cup_input.lower() in middle.lower():
            cup_choice = 2
        elif player_stone_cup_input.lower() in right.lower():
            cup_choice = 3
        else:
            print(f"Time is running, {red_text}{character_name.capitalize()}{reset}!! You sure you want to become a disciple?\n\nHint: please use numbers to proceed")

        # Creating a random cup with the spark 
        cup_with_spark = random.randint(1,3)

        #****** Progression condition ******
        if cup_with_spark == cup_choice: 
            print("You found the right cup, spark appeared, and the old door started to open with a cracking noise.")
            sleep(3)
            print(f"{'-'*117}")
            garden()

        #Repeating the while loop
        else:
            print("There was no spark inside the cup, maybe you should try again")
            print(f"\nThe spark was under the {cup_with_spark} cup")
            sleep(1)
            chances -=1
            if chances == 0:
                print(f"You have {chances} chances left.")

            else:
                print(f"You have {chances} chances left, please use your third sight\n")
                print("Stone cups began to mix again")
                print(f"{'-'*117}")
                sleep(2)
                continue

    # Fail condition
    else: 
        fail()
        
    
    
    
def garden():
    
    #Next room with the second progression game 
    
    #Describing the location 
    print("""\nIn front of you appeared a beautiful garden with a beautiful stream of unusual red color. 
However, your attention was drawn to the old and dry tree in the center of the garden.""")
    sleep(3)
    print('\n"So strange, everything is blooming around, but only this tree is dying"  - you thought.')
    sleep(3)
    print(f'{highlighted}\nEverything shrank inside you, feeling of pity and a passionate desire to help appeared in your soul, but how?{reset}')
    sleep(3)
    print('\nSuddenly, a Ghost lady appeared in front of you.')
    sleep(3)
    print(f'{highlighted}\n\nYou felt a cold shiver from the ghost,\nshe emanated hatred and anger, \nher skin was pale, \nonly her long, black hair stood out from her pale background.')
    sleep(3)
    print('\n\n"Pray for mercy. I feel your pity for this tree, but you will not help it!  \nIt belongs to me",\t- she whispered.')
    sleep(3)
    print('''\nThe Ghost lady feeds on the soul of this great tree, but she wasn't mind catching of your human soul too.
\nFear gripped your mind, your body numb... \n\t\tis this the end?''')
    sleep(3)
    print(f'The Ghost lady rushed to grab your throat, but suddenly your medallion was caught in a bright fire  and the shockwave pushed the Ghost lady  away.{reset}')
    sleep(3)
    print('''\n"This medallion has such a great power!
But I can't use it. \nI need to come up with something to stay alive.“ 
    \t\t\t\t\t\t- appeared in your thoughts.''')
    sleep(3)
    
    #Game introduction 
    print(f'''{'-'*117}\n\nYou asked the Ghost lady:\n
"I offer you a bet. If you win, you will get my soul, but if you lose,
you will let me go and breathe life  back into the tree."

Ghost lady couldn\'t resist the power of medallion, so she agreed to the bet.
\n
You found a couple of dice in your pocket, so you decided to try your luck with them.
{highlighted}You passed one dice to the ghost lady, and prepared for the game. 
Ghost lady, magically took the dice, and fed it with her powers by whispering something to the dice. 
\n
\nSomething is going wrong...{reset}''')
 
    #####################################################
    #
    # Second game with the dice random throw
    #
    #####################################################
    chances = 6
    while chances > 0:
        input(prompt="Press enter to throw the dice\n")
        #Higher roll for ghost due to the storyline
        ghost_throw = randint(6,6)

        #It was noted that 6 in the dice is very rare, therefore this if loop helps 
        #to proceed with the story by increasing the probability of throwing correct dice.
        if chances < 2:
            user_throw = randint(4,6)
        elif chances <= 5:
            user_throw = randint(2,6)
        else:
            user_throw = randint(1,6)

        print(f"Your dice showed {user_throw}, while the ghost throw {ghost_throw}.\n")
        sleep(2)

        #if loop for game progression/failure
        
            #****** Progression condition ******
        if user_throw == ghost_throw:
            print("The ghost lady screamed: 'You are the god killer' disappering in the scarlet light of your amulet.")
            sleep(2)
            #Accesing the final stage of the game 
            print(f"{'-'*117}")
            game_end()

        else:
            
            #Interactive response of ghost based on the user throw
            for number in range(user_throw):
                print("Ha ha")
            chances -=1
            if not chances == 0:
                print(f"\nPoor little child, I'll give you another try, you have {chances} chances left.")
            else: 
                print(f"You have {chances} chances left.")

    else:
        sleep(2)
        print("\nYou are going to die here, no amulet will save such weak donkey as you! I will feed on your soul.")
        fail()
        

    
def game_end():
    #Entering the final stage of the game
    sleep(5)
    print("""\nThe Ghost lady was defeated and disappeared into a thick fog.
\nIn the same moment, the old tree blossomed pointing with its branches at a newly opened door with a room behind.""")
    sleep(2)
    print(f"""\n
You walked into the room, and saw lots of books,
apparently, it was a library!\nIn the center of the room there was a chest decorated with stones.

Lost tome?

You rushed to the chest to open it... but the chest was empty... 

You noticed a small sign in the bottom of the chest, \nit was {opposite_clan} sign.

An ancient tome has been stolen from the library... 
Now you will have a long journey to find it. But that's a completely different story ...


To be continued....""")
    input(prompt="Press enter to exit\n")
    exit(1)
        
    

        
#Creating restart/exit function
def fail():
    reply = (input(prompt = "\nYou failed your quest, you cannot join the clan!!!\n\nDo you want to try again?[yes/no]")).lower().strip()
             
    if reply[0] == 'y':
        print(f'''{red_text}{character_name.capitalize()}{reset}, let your journey start again''')
        sleep(1)
        game_start()
    elif reply[0] == 'n':
        print(f'\n\nThank you for your time, {red_text}{character_name.capitalize()}{reset}!\nStay Safe \U0001F637\U0001F637\U0001F637!!!')
        exit(0)
    #Debugging         
    else:
        print(
            f'Sorry, {red_text}{character_name}{reset}, you did something wrong!\nPlease try again!')
        input(prompt='<Press enter to proceed>\n')
        fail()
```
